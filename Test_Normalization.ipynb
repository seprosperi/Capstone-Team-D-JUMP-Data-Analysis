```python
import pandas as pd 
import fastparquet #make sure that this is installed in your anaconda environment
import numpy as np
import statistics as stat
import matplotlib.pyplot as plt

#loads the file into the label "data"
filename='1053600674.parquet' #edit the filename here for testing other .parquet files

data=pd.read_parquet(filename, engine='fastparquet') 

data_numerical=data._get_numeric_data() #removing columns that have non quantitiative data
```

```python
from scipy import stats

#this is what I did on my own

meds=data_numerical.median(axis=0) #find the median

MADs=stats.median_abs_deviation(data_numerical,axis=0) #is this correct? find the MAD for each column

Zs=(0.6745*(data_numerical-meds))/MADs #implement robust Z transformation

Zs #display the results
```

```python
#retest this with ryan's code to make sure its the same

df_normalized = (0.6745 * (data_numerical - data_numerical.median(axis=0))) / stats.median_abs_deviation(data_numerical, axis=0)

df_normalized #hell yeah it looks the same
```

```python
np.abs(df_normalized)
```

```python
#remove outliers
#this is also taken right from Ryan's slides

n_MAD = 500
threshold = 0.6745 * n_MAD
rows_to_keep = (np.abs(df_normalized) < threshold).all(axis=1)
df_normalized = df_normalized[rows_to_keep]

df_normalized #hmmm so where did the data go?

#add in the metadata
```
