```python
parquet_file_path = '/Users/gaiadibernardini/Downloads/1086293072.parquet'

# Read the Parquet file into a DataFrame
df = pd.read_parquet(parquet_file_path, engine='fastparquet')
df_numerical=df._get_numeric_data()
```

```python
median = np.median(df_numerical)
absolute_deviation = np.abs(df_numerical - median)
MAD = np.median(absolute_deviation)
df_normalized = (0.6745 * (df_numerical - df_numerical.median(axis=0))) / MAD
df_normalized.shape
display(df_normalized)
```

```python
df_normalized = (0.6745 * (df_numerical - df_numerical.median(axis=0))) / stats.median_abs_deviation(df_numerical, axis=0)
df_normalized = df_normalized.dropna(axis=1) #dropping NaN values from columns

n_MAD = 500
threshold = 0.6745 * n_MAD
rows_to_keep = (np.abs(df_normalized) < threshold).all(axis=1)
df_normalized = df_normalized[rows_to_keep]

df_normalized
```
